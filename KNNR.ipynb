```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('MLDataset.csv')
dataS = data.drop('Unnamed: 0',axis=1)
```

```python
dataS
```

```python
scaler = StandardScaler()
scaler.fit(dataS)
dataS = scaler.transform(dataS)
```

```python
X = dataS[:,0:3]
Y = dataS[:,3:]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.15,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Finding the Best Degree
Rscore = {}
for i in range(2,200):
    Rscore[i]={}
    for j in ['uniform', 'distance']:
        Model = KNeighborsRegressor(n_neighbors=i,weights=j)
        Model.fit(X_train,Y_train[:,0])
        y_pred = Model.predict(X_test)
        val = Model.score(X_test,Y_test[:,0])
        Rscore[i][j]=val
```

```python
Model = KNeighborsRegressor(n_neighbors=22,weights='uniform')
Model.fit(X_train,Y_train[:,0])
y_pred = Model.predict(X_test)
```

```python
Rscore
```

```python
y_pred = y_pred.reshape(y_pred.shape[0],1)
```

```python
arr = np.concatenate((X_test,y_pred,Y_test[:,1:]),axis=1)
df = pd.DataFrame(arr)
A = scaler.inverse_transform(df)
```

```python
arr = np.concatenate((X_test,Y_test),axis=1)
df = pd.DataFrame(arr)
B = scaler.inverse_transform(df)
```

```python
val = 0
for i in range(A.shape[0]):
    error = abs((A[i,-4]-B[i,-4])/B[i,-4])
    val+=error
```
