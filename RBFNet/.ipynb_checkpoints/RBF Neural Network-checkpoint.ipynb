```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from RBFLayer import RBFLayer,KMeansCluster
from keras.layers import Dense,Activation
from keras.models import Sequential
```

```python
data = pd.read_csv("../Dataset/PCC/MLDataset_quat.csv")
dataS = data.drop('Unnamed: 0',axis=1)
scaler = MinMaxScaler(feature_range=(0,1))
scaler.fit(dataS)
dataS = scaler.transform(dataS)
X = dataS[:,0:7]
Y = dataS[:,7:]
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=0)
```

```python
model = Sequential()
model.add(RBFLayer(output_dim=17,initializer=KMeansCluster(X_train),betas=0.1))
model.add(Dense(Y.shape[1],activation='sigmoid'))
```

```python
model.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
model.fit(X_train,Y_train, epochs=150, batch_size=64)
```
